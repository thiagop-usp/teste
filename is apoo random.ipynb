```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd
import seaborn as sns
import os
```

```python
turma1 = pd.read_csv('../input/turma1-apoo-notas/turma1.csv')
turma2 = pd.read_csv('../input/turma-2-notas-apoo/turma2.csv')
```

```python
turma1 = turma1.drop(columns='Unnamed: 1')
turma2 = turma2.drop(columns='Unnamed: 1')
```

```python
turma1.head()
```

```python
turma2.head()
```

```python
turma1_m = turma1.mean()
turma1_st = turma1.std()
turma2_m = turma2.mean()
turma2_st = turma2.std()
```

```python
from scipy.stats import norm
```

```python
sns.distplot(turma1, label='turma1', color='blue')
sns.distplot(norm.rvs(loc=turma1_m, scale=turma1_st, size=100), label='normal', color='orange')
```

```python
sns.distplot(turma1, label='turma2', color='blue')
sns.distplot(norm.rvs(loc=turma2_m, scale=turma2_st, size=100), label='normal', color='orange')
```

```python
tudo = pd.concat([turma1['Nota'], turma2['Nota']], axis=0)
```

```python
tudo_m = tudo.mean()
tudo_st = tudo.std()

sns.distplot(turma1, label='tudo', color='blue')
sns.distplot(norm.rvs(loc=tudo_m, scale=tudo_st, size=100), label='normal', color='orange')
```
